# Goal Scored And Conceeded

```python
# Filter Team And Goals Scored
asnl_24_25_data = asnl_24_25_data[(asnl_24_25_data['Team'] == 'Arsenal') & (asnl_24_25_data['Gls'] >= 0)]

# Show Df
asnl_24_25_data
```

```python
# Create Size Of The Plot
plt.figure(figsize = (11, 7))

# Add Fixtures To Xlabel
fixtures = asnl_24_25_data['Fixture']

# Create Axes Variable
ax = plt.axes()

# Set The Background Color (Hex Number)
ax.set_facecolor('#77a1b5')

# Plot Lines. Use *.index* For X Axis, Call Df Then Column For Y Axis, Add Shape, Line Type And Color, Choose Marker Size (ms) And Marker Edge Color (mec), Label Name For Legend 
plt.plot(asnl_24_25_data.index, asnl_24_25_data['Gls'],'o--r', ms=11, mec='w', label ='Goals For')
plt.plot(asnl_24_25_data.index, asnl_24_25_data['Sht Stop GA'],'o--b', ms=11, mec='w', label ='Goals Against')

# Name Axis labels And Title And Change Font Weight
plt.xlabel('Opposition', fontweight='bold')
plt.ylabel('Goals And Goals Against', fontweight='bold')
plt.title('24/25 Season - Goals After 11 Arsenal Games', fontweight='bold')

# Plot x-ticks Using Index Column And Label With Fixture Variable, Rotate And Move 
plt.xticks(ticks=asnl_24_25_data.index, labels=fixtures, rotation=45, ha='right')

# Plot Legend And Choose Location
plt.legend(loc = "upper left")

# Show Grid Lines
plt.grid(True, color='black')
```

```python
# Store File In Variable, Read File 
comb_gls = pd.read_csv('combined_goals.csv')

# Show Df
comb_gls
```

```python
# Create New Columns =   From Df Choose Column & Use Cumulative Sum Function x2
comb_gls['Cumulative Goals 23-24'] = comb_gls['Goals 23-24'].cumsum()
comb_gls['Cumulative Goals 24-25'] = comb_gls['Goals 24-25'].cumsum()

# Plt Graph
    # Figure(figsize())
    # Plot(From Df Select Column *'Game Week'*, Make New Columns, Marker(Marker Point = 'o', Line Style = '--', Color = 'r') Label For Legend)
plt.figure(figsize=(10,6))
plt.plot(comb_gls['Game Week'], comb_gls['Cumulative Goals 23-24'], 'o--r', label='Goals 23-24')
plt.plot(comb_gls['Game Week'], comb_gls['Cumulative Goals 24-25'], 'o--b', label='Goals 24-25')
    
# Plot Labels & Tiles, Change *fontweight* To Bolden Text
plt.title('Comparison Of Arsenal Goals Scored After 11 Games', fontweight='bold')
plt.xlabel('Game Week', fontweight='bold')
plt.ylabel('Number Of Goals Scored', fontweight='bold')

#'Get Current Axis' *gca()*  And *set_facecolor* To Change Background
plt.gca().set_facecolor('#77a1b5')

# Rotate The X-Axis Labels And Move 'right' To Fit Under Bars 
plt.xticks(rotation=45, ha='right')

# Plot Legend (Auto Location Top Left)
plt.legend()

# Show Grid Lines (Only y-axis 'Horizontal')
plt.grid(axis='y', color='black')

# Displays Figure
plt.show()
```

# 24/25 Times Of Goals

```python
with open('mins_scored_quaters.csv', 'r') as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        print(row)
```

```python
# Create Variable To Store The csv Data. Using read_csv Function
quarter = pd.read_csv('mins_scored_quaters.csv')

# Using Same Varibale Set The Index For Df
quarter = quarter.set_index('Game Week')

# Call Variable To Show Df
quarter
```

```python
# Calculate Each Column Sum
# Create Variable, Use iloc Fucntion To Slice Columns, Then Sum The Results
    #[:,] = All Column
    #[, 1:] = From The First Row And Everything After
column_mins = quarter.iloc[:, 1:].sum()

# Create A New Df For Data
# Create Variable, Use Pandas Df Function, Using The *column_mins* df, Name The Column For The Minutes 
quarter_df = pd.DataFrame(column_mins, columns=['Goals'])

# Reset Index Back To Original(0,1,2), Previous Index Becomes 1st Column
quarter_df.reset_index(inplace=True)

# Change Column Names For Clarity
quarter_df.columns = ['Mins','Goals']
quarter_df
```

```python
# Create Bar Graph Using *figure* Function And Determining Size Of The Graph With *figsize*
plt.figure(figsize=(10, 7))

# Plot Bar Select Columns Needed in *[]*, Change Bar *color*
plt.bar(quarter_df['Mins'], quarter_df['Goals'], color='maroon')

# Plot Labels And Title, Change Fontsize And Weight 
plt.xlabel('Minutes', fontsize=12, fontweight='bold')
plt.ylabel('Goals', fontsize=12, fontweight='bold')
plt.title('Minutes Arsenal Scored 24/25', fontsize=14, fontweight='bold')

# Rotate The X-Axis Labels And Move 'right' To Fit Under Bars 
plt.xticks(rotation=45, ha='right')

# Set Background Color With *gca()* 'Get Current Axes* Function 
plt.gca().set_facecolor('#77a1b5')

# Adjust Spacing For Graph
plt.tight_layout()

# Displays Figure
plt.show()
```

```python
# Create Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 0 - 22.5','Min 22.5 - 45','45 +','Min 45 - 67.5','Min 67.5 - 90','90+'],
     'Total Goals': [3,3,2,4,4,2]}
  
# Create a Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','Total Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['Total Goal Percentage'] = (goal_mins_df['Total Goals'] /
                   goal_mins_df['Total Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```

```python
# Create First Half Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 0 - 22.5','Min 22.5 - 45','45 +'],
     'First Half Goals': [3,3,2]}
  
# Convert To A Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','First Half Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['First Half Goal Percentage'] = (goal_mins_df['First Half Goals'] /
                   goal_mins_df['First Half Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```

```python
# Create Second Half Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 45 - 67.5','Min 67.5 - 90','90+'],
     'Second Half Goals': [4,4,2]}
  
# Convert To A Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','Second Half Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['Total Goal Percentage'] = (goal_mins_df['Second Half Goals'] /
                   goal_mins_df['Second Half Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```
