```python
# Create Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 0 - 22.5','Min 22.5 - 45','45 +','Min 45 - 67.5','Min 67.5 - 90','90+'],
     'Total Goals': [3,3,2,4,4,2]}
  
# Create a Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','Total Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['Total Goal Percentage'] = (goal_mins_df['Total Goals'] /
                   goal_mins_df['Total Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```

```python
# Create First Half Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 0 - 22.5','Min 22.5 - 45','45 +'],
     'First Half Goals': [3,3,2]}
  
# Convert To A Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','First Half Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['First Half Goal Percentage'] = (goal_mins_df['First Half Goals'] /
                   goal_mins_df['First Half Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```

```python
# Create Second Half Dictionary To Store Data For Df
goal_mins_dict = {
     'Minute': ['Min 45 - 67.5','Min 67.5 - 90','90+'],
     'Second Half Goals': [4,4,2]}
  
# Convert To A Df Create Column Names
goal_mins_df = pd.DataFrame(goal_mins_dict, 
                   columns = ['Minute','Second Half Goals'])
 
# Calculating Percentage For The Data
goal_mins_df['Total Goal Percentage'] = (goal_mins_df['Second Half Goals'] /
                   goal_mins_df['Second Half Goals'].sum()) * 100
 
# Show Df, Set Index, Data Rounded To 1 Decimal Place
goal_mins_df = goal_mins_df.set_index('Minute')
goal_mins_df.round(1)
```
